### To use this API, just enter the features starting from the 4th argument of ```encode.FeatureConfig```. Feature names should be exact same name as the feature in the list above 

#### if you know exact tissue ontology available in ENCODE (2nd arg should be True; if you pass 3rd arg as False it will search for tissue samples):

````python
c = encode.FeatureConfig('prostate gland',  True, False, "H3K4me3", "DNase", "POLR2A", "CTCF", "transcription")
results = c.parse_url(rfa='ENCODE4') # you can change it to ENCODE3, to restrict files processed from one of ENCODE3 or ENCODE4
results
````
